```python
import re
from collections import Counter
from pathlib import Path
```

```python
def calculate_password_score(password):

    # Rule 1: Base score is password length
    base_score = len(password)

    # Rule 2: Required character types
    has_lowercase = bool(re.search(r'[a-z]', password))
    has_uppercase = bool(re.search(r'[A-Z]', password))
    has_digit = bool(re.search(r'[0-9]', password))
    has_special = bool(re.search(r'[!@#$%^&*]', password))

    missing_categories = sum([
        not has_lowercase,
        not has_uppercase,
        not has_digit,
        not has_special
    ])

    score = base_score * (0.75 ** missing_categories)

    # Rule 3: Repeated characters (30% threshold)
    char_counts = Counter(password)
    most_common_char, max_count = char_counts.most_common(1)[0]

    if max_count >= len(password) * 0.3:
        score -= max_count

    return score


dataset_path = Path("../datasets/day03/hyperskill-dataset-119011491.txt")

with open(dataset_path, "r") as f:
    passwords = [line.strip() for line in f if line.strip()]

# ðŸ” Analyze passwords
best_password = None
best_score = -float('inf')

print("Password Analysis")
print("=" * 80)

for password in passwords:
    score = calculate_password_score(password)
    print(f"{password:30s} -> Score: {score:.2f}")

    # Tie-breaking: earlier password wins automatically
    if score > best_score:
        best_score = score
        best_password = password

print("=" * 80)
print(f"Best Password : {best_password}")
print(f"Security Score: {best_score:.2f}")
```
