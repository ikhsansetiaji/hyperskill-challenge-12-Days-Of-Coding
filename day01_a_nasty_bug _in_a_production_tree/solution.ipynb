```python
from collections import Counter

with open("../datasets/day01/hyperskill-dataset-118968836.txt", "r") as f:
    lines = f.readlines()

all_errors = []
incident_errors = []

for line in lines:
    line = line.strip()
    if not line:
        continue

    time, error = line.split()
    all_errors.append(error)

    if time.startswith("15:"):
        minute = int(time.split(":")[1])
        if minute <= 30:
            incident_errors.append(error)

background_noise = Counter(all_errors).most_common(1)[0][0]
incident_counter = Counter(incident_errors)

culprit = None
for err, count in incident_counter.most_common():
    if err != background_noise:
        culprit = err
        break

print("Background noise error:", background_noise)
print("Incident culprit:", culprit)
print("Count:", incident_counter.get(culprit, 0))
```
